```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, binom, poisson, expon

sns.set(style="whitegrid")

# Set up the plot grid
fig, axs = plt.subplots(4, 2, figsize=(14, 16))
fig.suptitle("Probability Distributions: PDF and CDF", fontsize=18, fontweight='bold')

# -------------------- Normal Distribution (e.g., dice roll average) --------------------
x = np.linspace(-4, 4, 1000)
mu, sigma = 0, 1  # Standard Normal

axs[0, 0].plot(x, norm.pdf(x, mu, sigma), color='blue')
axs[0, 0].set_title("Normal Distribution - PDF")
axs[0, 1].plot(x, norm.cdf(x, mu, sigma), color='green')
axs[0, 1].set_title("Normal Distribution - CDF")

# -------------------- Binomial Distribution (e.g., success in 10 dice rolls) --------------------
n, p = 10, 1/6  # 10 dice rolls, success = rolling a 6
x_binom = np.arange(0, n+1)

axs[1, 0].bar(x_binom, binom.pmf(x_binom, n, p), color='purple')
axs[1, 0].set_title("Binomial Distribution - PMF (10 rolls, p=1/6)")
axs[1, 1].plot(x_binom, binom.cdf(x_binom, n, p), marker='o', color='orange')
axs[1, 1].set_title("Binomial Distribution - CDF")

# -------------------- Poisson Distribution (e.g., customers arriving per minute) --------------------
λ = 4  # average 4 customers/min
x_poisson = np.arange(0, 15)

axs[2, 0].bar(x_poisson, poisson.pmf(x_poisson, λ), color='red')
axs[2, 0].set_title("Poisson Distribution - PMF (λ = 4)")
axs[2, 1].plot(x_poisson, poisson.cdf(x_poisson, λ), marker='o', color='brown')
axs[2, 1].set_title("Poisson Distribution - CDF")

# -------------------- Exponential Distribution (e.g., time between arrivals) --------------------
x_exp = np.linspace(0, 10, 1000)
rate = 1  # λ = 1 -> mean = 1

axs[3, 0].plot(x_exp, expon.pdf(x_exp, scale=1/rate), color='teal')
axs[3, 0].set_title("Exponential Distribution - PDF (λ = 1)")
axs[3, 1].plot(x_exp, expon.cdf(x_exp, scale=1/rate), color='magenta')
axs[3, 1].set_title("Exponential Distribution - CDF")

# Label all axes
for ax in axs.flat:
    ax.set_xlabel("x")
    ax.set_ylabel("Probability")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
